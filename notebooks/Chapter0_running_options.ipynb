### Running on your own computer

There are many options you can use for running the course notes on your onwn computer. These all need network access *at some point*, so you can download the files you need. If you do this, you are responsible for your own backups, so might want to use some servive such as google drive, dropbox etc. for backup. These would need only intermittant network access (i.e. not all of the time you are working). You could always sort yopur owen backup on local respources if network access is a real problem.

#### docker

In many ways, the simplest way to set up and run the notes on your own computer is to use [`docker`](https://www.docker.com/products/docker-desktop). First doenload and install the docker software.

Then download the docker file [`proflewis/geog0111:latest`](https://hub.docker.com/repository/docker/proflewis/geog0111) using the command line:

    docker pull proflewis/geog0111:latest
    
or using the Docker desktop app.

You can then run jupyter lab with:

    docker run -it --rm -p 8888:8888 proflewis/geog0111:latest start.sh jupyter lab
    
This will point you to a url you can open in a browser, e.g. [http://localhost:8888](http://localhost:8888/?token=d710ebf6d9d26ccfe375d083d4e6ca86babea99410e0a9d5).

Note that you can only have one jupyter session running on your computer at a time.

Navigate to the notebook for the session (e.g. `notebooks/Chapter0_help.ipynb`) and save any work you do. It will be there in the docker jupyter hub the next time you start it.

You quit an `jupyter` notebook session with `^C` (`Control C`).

#### anaconda and git

The most flexible is to install [Anaconda python](https://docs.anaconda.com/anaconda/install/) on your local computer.

Once you have done that, you can use `conda` at the command-line, or tools such as [`anaconda-navigator`](https://docs.anaconda.com/anaconda/navigator/) to install packages.

You will need to install [`git`](http://git-scm.com/) and `pip` as a minimum, to allow other mechanisms to download packages.

To keep track of files in `git`, you can also use various `git` desktop applications such as [`git desktop`](https://desktop.github.com/). 

So, we should now have `conda`, `pip` and `git` installed on the local computer. Yopu can test this by typing the following in a terminal:

    which git conda pip
    
and this should respond with something like:

    /Users/jovian/anaconda3/envs/newform0111/bin/git
    /Users/jovian/anaconda3/envs/newform0111/bin/conda
    /Users/jovian/anaconda3/envs/newform0111/bin/pip

If it doesn't respond with an appropriate path to these commands, try restarting your computer or session to see if that has updated things. If not, try installing them again.

Once we know we have these, you can pull the course repository from github, typing:

    git clone https://github.com/profLewis/newform0111.git
    
or using github desktop.

To install these:

    cd newform0111
    conda env create -f environment.yml
    
If you are updating the notes, and newform0111 already exists, use instead:

    cd newform0111
    conda env update -f environment.yml


This will create an environment called `newform0111` and make sure you have all of the required dependencies.

If you have created the environment, you can activate it with:


    conda activate newform0111

and finally, set up code and notebook features, using:

    cd newform0111
    python setup.py install
    export PATH="${HOME}/.local/bin/:${PATH}" && bash postBuild

For further advice on checking, setting or deleting `conda` environments, see the [conda help pages](https://conda.io/docs/user-guide/tasks/manage-environments.html).

To run the Jupyter notebook server, simply type:

    jupyter notebook
    
and navigate to the notebook  for the session (e.g. `notebooks/Chapter0_help.ipynb`).

You quit an `jupyter` notebook session with `^C` (`Control C`).

To exectute ('run') blocks of Python code in the notebook, use `^<return>` (`SHIFT` and `RETURN` keys together).

Alternatively, just run `ipython`:  
```
cd /directory/to/newform0111
ipython
```

and type your own commands in at the prompt, following the class or the material on the webpages.


### Hybrid approaches

If you have a fast local computer, but only intermittant network access, you might like to go for a hybrid approach. For example, once you set up on google colab, you can save your notes and codes to your google drive, which you can have mirrored to your local computer.

Alternatively, you could set up some online repository such as in github, and you can upload your work there once you have finished a session. If you do this, make sure it is not public though, as the course includes work you must submit.
